```python
import matplotlib.pyplot as plt
plt.plot([2,4,6,4])
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.0,0.2)

#blue dashes, red squares and green triangles
plt.plot(t,t**2,'b--',label='^2') # 'rs' 'g^')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.2,'g',label='^2.5')
plt.grid()
plt.legend() #add legend based on line labels
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y, linewidth=5.0)
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)

#use keyword args
plt.setp(lines[0],color='r',linewidth=2.2)

#or MATLAB style string value pairs
plt.setp(lines[1],'color','g','linewidth',2.0)


plt.grid()
```
